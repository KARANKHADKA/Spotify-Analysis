### Dataset Features
The features in our dataset are described in the below. These descriptions are taken from the Kaggle data source.

1. track_name: Name of the song
2. artist(s)_name: Name of the artist(s) of the song
3. artist_count: Number of artists contributing to the song
4. released_year: Year when the song was released
5. released_month: Month when the song was released
6. released_day: Day of the month when the song was released
7. in_spotify_playlists: Number of Spotify playlists the song is included in
8. in_spotify_charts: Presence and rank of the song on Spotify charts
9. streams: Total number of streams on Spotify
10. in_apple_playlists: Number of Apple Music playlists the song is included in
11. in_apple_charts: Presence and rank of the song on Apple Music charts
12. in_deezer_playlists: Number of Deezer playlists the song is included in
13. in_deezer_charts: Presence and rank of the song on Deezer charts
14. in_shazam_charts: Presence and rank of the song on Shazam charts
15. bpm: Beats per minute, a measure of song tempo
16. key: Key of the song
17. mode: Mode of the song (major or minor)
18. danceability_%: Percentage indicating how suitable the song is for dancing
19. valence_%: Positivity of the song's musical content
20. energy_%: Perceived energy level of the song
21. acousticness_%: Amount of acoustic sound in the song
22. instrumentalness_%: Amount of instrumental content in the song
23. liveness_%: Presence of live performance elements
24. speechiness_%: Amount of spoken words in the song

## Data Cleaning/ Data Wrangling

```python
# Checking for missing data
for col in spotify_df.columns:
    dmp = spotify_df[col].isnull().sum()
    print(f'{col} --> {dmp*100}')
```

```python
#Dropping all missing values and checking again
spotify_df = spotify_df.dropna()

for col in spotify_df.columns:
    dmp = np.mean(spotify_df[col].isnull())
    print(f'{col} --> {dmp*100}%')
```

```python
# Check for the data types of all  the coloumns
print(f"Data types are below where 'object' indicates a string type: ")
print(spotify_df.dtypes)
```

```python
# the data types do not match the intended data types, thus we have to change the datatypes for some coloumns.
spotify_df['streams'] = spotify_df['streams'].astype(str).str.extract('(\d+)', expand=False)
spotify_df['streams'] = pd.to_numeric(spotify_df['streams'], errors='coerce')

spotify_df['in_deezer_playlists'] = spotify_df['in_deezer_playlists'].astype(str).str.extract('(\d+)', expand=False)
spotify_df['in_deezer_playlists'] = pd.to_numeric(spotify_df['in_deezer_playlists'], errors='coerce')

spotify_df['in_shazam_charts'] = spotify_df['in_shazam_charts'].astype(str).str.extract('(\d+)', expand=False)
spotify_df['in_shazam_charts'] = pd.to_numeric(spotify_df['in_shazam_charts'], errors='coerce')

spotify_df['streams'] = spotify_df['streams'].astype('int64')
spotify_df['in_deezer_playlists'] = spotify_df['in_deezer_playlists'].astype('int64')
spotify_df['in_shazam_charts'] = spotify_df['in_shazam_charts'].astype('float64')
```

```python
# Check for data types
print(f"Data types are below where 'object' indicates a string type: ")
print(spotify_df.dtypes)
```

```python
# Joining columns, key and mode to make one column called chord and dropping those coloumns
spotify_df["chord"] = spotify_df["key"] + spotify_df["mode"]
spotify_df = spotify_df.drop(columns=['key', 'mode'])
```

## Data Exploration and Visualisation

Our dataset is now considered to be clean and we are ready to start visualising and exploring the features of the dataset.

### Two Plots Displaying the Distribution of a Single Continuous Variable

```python
plt.figure(figsize=(20, 10))
fig = sns.histplot(spotify_df['energy_%'], kde=True,color='blue').set_title('Figure 1: Histogram of Energy level in songs', fontsize = 20)
plt.xlabel('Energy level of songs(%)',fontsize=20)
plt.ylabel('Number of songs',fontsize=20)
caption = "The Histogram shows the distribution of songs for different energy levels"
plt.text(0.5, -0.15, caption, horizontalalignment='center', verticalalignment='center', 
         fontsize=20, transform=plt.gca().transAxes)
plt.show();
```

#### The Histogram of Energy level in songs, is left-skewed, uni-modal data. Majority of the songs lie between energy levels of 60%-80%.

```python
plt.figure(figsize = (20,10))
sns.boxplot(spotify_df['bpm']).set_title('Figure 2: Box Plot for the range of bpm of tracks', fontsize = 15)
plt.ylabel('BPM of tracks',fontsize=20)
caption = "The boxplot gives the description of the range of bpm of tracks"
plt.text(0.5, -0.10, caption, horizontalalignment='center', verticalalignment='center', 
         fontsize=20, transform=plt.gca().transAxes)
plt.show();
```

#### The boxplot shows the distribution of songs based on the bpm of tracks. 
#### Inter-Quartile Range(IQR) = Q3 - Q1 
#### IQR= 141 - 99 = 42
#### Median = 120

### Two Plots Displaying Information about a Single Categorical Variable

```python
plt.figure(figsize=(25, 15))
fig = sns.countplot(y='chord', data=spotify_df, palette="blend:#7AB,#EDA",
                    order=spotify_df['chord'].value_counts().index,orient='h')
fig.set_title('Figure 3: Most frequently used Chords', fontsize=20)
plt.xlabel('Frequency',fontsize=20)
plt.ylabel('Chords',fontsize=20)
caption = "The Horizontal Bar chart displays the number of songs in the dataset that use a particular chord"
plt.text(0.5, -0.15, caption, horizontalalignment='center', verticalalignment='center', 
         fontsize=20, transform=plt.gca().transAxes)
plt.show()
```

#### The horizontal bar plot shows us that 'C# Major' chord is used in maximum number of songs followed by 'DMajor' and 'GMajor'.
#### 'D#Major' is the chord that is used in the least number of songs

```python
plt.figure(figsize=(20, 10))
fig = sns.countplot(x='released_month', data=spotify_df, palette='magma',
                    order=spotify_df['released_month'].value_counts().index)
fig.set_title('Figure 4 : Number of releases every month', fontsize=20)
plt.xlabel('Month',fontsize=20)
plt.ylabel('Number of releases',fontsize=20)

# Annotate the bars with the counts
for p in fig.patches:
    fig.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')
    
caption = "The Bar chart displays the number of releases in the dataset from each month"
plt.text(0.5, -0.15, caption, horizontalalignment='center', verticalalignment='center', 
         fontsize=20, transform=plt.gca().transAxes)
plt.show()
```

#### The barplot shows us that the maximum number of songs in this dataset are released in May, followed by January. August has the least number of releases in this dataset.

### One Plot Displaying Information about Both a Continuous Variable and a Categorical Variable

```python
plt.figure(figsize=(20, 10))
top_10_artists = spotify_df.groupby('artist(s)_name')['streams'].sum().nlargest(10).index
groupby_data = spotify_df[spotify_df['artist(s)_name'].isin(top_10_artists)]
fig = sns.barplot(data=groupby_data, x='artist(s)_name', y='streams', order=top_10_artists,ci=None)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Top 10 Artists',fontsize=20)
plt.ylabel('Streams',fontsize=20)
plt.title('Figure 5: Top 10 Artists with Highest Streams(in Millions)',fontsize=20)
for p in fig.patches:
    fig.annotate(f'{p.get_height() / 1e6:.1f} M', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

caption = "The Bar chart displays the number of streams for the Top 10 artists in Millions"
plt.text(0.5, -0.25, caption, horizontalalignment='center', verticalalignment='center', 
         fontsize=20, transform=plt.gca().transAxes)

plt.show()
```

#### We can deduce from this barplot that Imagine dragon has the most number of streams with 1757.5 Millions followed by Ed Sheeran and Eminem with 1578.8 Millions 1546 Millions respectively.

### Two Plots Showing a Relationship Between Two Variables

```python
plt.figure(figsize=(20, 10))
sns.lineplot(data=spotify_df[spotify_df['released_year'] > 2000], x='released_year',
             y=spotify_df['bpm'],ci=None)
# Customize the chart
plt.xlabel('Year of release (2000-2023)',fontsize=20)
plt.ylabel('BPM',fontsize=20)
plt.title('Figure 6: Relationship between BPM and year of release.',fontsize=20)

caption = "The Lineplot displays relationship between the bpm of the tracks and the year of release(2000-2023)"
plt.text(0.5, -0.15, caption, horizontalalignment='center', verticalalignment='center', 
         fontsize=20, transform=plt.gca().transAxes)

plt.show()
```

#### The lineplot displays that the BPM of tracks goes down from around 140 bpm to a little over 90 bpm from 2000 to 2005.
#### Then the BPM of songs goes above 150 between 2010 and 2015.
#### The BPM of songs has remained between 110 bpm to 130 bpm from 2015.

```python
plt.figure(figsize=(20, 10))
sns.boxplot(x='released_month', y='streams', data=spotify_df, palette="Set3")
plt.title("Figure 7: Streams by released month",fontsize=20)
plt.xlabel("Released month",fontsize=20)
plt.ylabel("Streams",fontsize=20)
plt.xticks(rotation=45)

caption = "Boxplot displaying relationship between release month by number of streams"
plt.text(0.5, -0.15, caption, horizontalalignment='center', verticalalignment='center', 
         fontsize=20, transform=plt.gca().transAxes)

plt.show()
```

#### The boxplot shows that the songs released in January have the maximum number of streams. 

### One plot should display information about 3 or more variables

```python
plt.figure(figsize=(30, 15))

top_5_songs = spotify_df.groupby('track_name')['streams'].sum().nlargest(5).index
groupby_top5 = spotify_df[spotify_df['track_name'].isin(top_5_songs)]
selected_columns = ['danceability_%', 'valence_%', 'energy_%', 'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%','bpm']

heatmap_data = groupby_top5.pivot_table(index='track_name', values=selected_columns)
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.2f', linewidths=1, linecolor='black', cbar_kws={'label': 'Values'}, annot_kws={'size': 15})

plt.title('Figure 8: Audio Features by Top 5 Songs',fontsize=20)
plt.xlabel('Audio Features',fontsize=20)
plt.ylabel('Songs',fontsize=20)

caption = "Heatmap displaying relationships between the audio features by top 5 songs"
plt.text(0.5, -0.10,caption, horizontalalignment='center', verticalalignment='center', 
         fontsize=20,transform=plt.gca().transAxes)

plt.show()
```

#### The heatmap shows the relationships between the audio features and the top 5 songs. We can conclude that the top 5 songs have a mid to high level range(90 to 170 bpm) for bpm, mid level range(50 to 83) for danceability and energy.  Acousticness ranges from a low to mid level range(1 to 58).  

### Questions and Answers

#### Q1.  In what ways do you think data visualization is important to understanding a data set? 
* Pattern Recognition: Data visualization simplifies pattern recognition in datasets by making trends more easy to recognize.
* Outlier Detection: Visualizations are instrumental in detecting outliers, ensuring data quality and integrity.
* Insights, Relationships and Correlations: They reveal valuable insights about relationships and correlations between variables, helping analysts make data-driven decisions.
* Summarization: Visualizations can summarize large amounts of data in a compact and understandable form.
* Communication: Data visualizations are an effective way to communicate data-driven insights to non-technical stakeholders.
* Storytelling: Visualization helps to tell stories by curating data into a form of narrative explanation by using graphs that are easier to understand.

#### Q2. In what ways do you think data visualization is important to communicating important aspects of a data set?
Data visualization is essential for effectively communicating important aspects of a dataset for several reasons:
* Visualizations simplify complex data, making it easier for a broad audience to grasp key insights.
* Visualizations emphasize patterns and trends, which can be vital in conveying the significance of data points or trends over time.
* They can be used to strengthen the narrative surrounding the data, providing a more compelling context for the audience.
* Visualizations aid decision-making by providing a clear and concise summary of relevant data, helping users make informed choices.
* Visualizations enable easy comparisons between different data points, helping users identify variations, similarities, and outliers.
* They condense a large amount of data into a concise format, allowing for efficient communication of the most important aspects.

#### Q3. What role does your integrity as an analyst play when creating a data visualization for communicating results to others?
* Accuracy: Integrity as an analyst is is to ensure that the data visualization accurately represents the data. 
* Transparency: An honest and transparent approach to data visualization involves clearly representing the data, including any limitations or uncertainties. This helps viewers understand the context and potential biases.
* Ethical Considerations: Ethical concerns, such as privacy and consent, are vital when working with data. Maintaining integrity means respecting these ethical considerations in data collection and visualization.
* Impact Assessment: Evaluating the potential impact of visualization choices on the audience and decision-making is an integral part of maintaining integrity.

#### Q4. How many variables do you think you can successfully represent in a visualization? What happens when you exceed this number?

* Generally, it's recommended to limit the number of variables to around 5-7 in a single visualization to maintain clarity.
* The visualization becomes overcrowded, making it difficult for viewers to discern patterns or draw meaningful insights.
* Too many variables can lead to confusion or misinterpretation of the data, which defeats the purpose of the visualization.
* Viewers may experience cognitive overload when trying to process excessive information, leading to fatigue and reduced comprehension.
* However to represent a larger number of variables, we might consider creating multiple smaller visualizations or interactive visualizations to allow users to explore the data selectively.
